```python
import numpy as np
import matplotlib.pyplot as plt
def calculate_six_parameter_fit(reaction,dictonary,temperature):
    #finish editing this 
    #calc Ea,c,d,F seprately 
        A = dictonary[reaction]['A']
        n = dictonary[reaction]['n']
        Ea_temp = dictonary[reaction]['Ea']/(1.987*temperature)
        c_temp = dictonary[reaction]['c']/((1.987*temperature)**3)
        d_temp = dictonary[reaction]['d']*(1.987*temperature)
        f_temp = dictonary[reaction]['f']* ((1.987*temperature)**3)

        k = A*(temperature**n)*np.exp(-Ea_temp-c_temp-d_temp-f_temp)
        return k 
```

```python
#plotting the difference between original msi and msi with new hong data as optimization targets 
def plotting_reaction(reaction,dictonary1,dictonary2):
    k1 = []
    k2=[]
    k_hong_cti  = []
    temperature = np.arange(200,2400)
    for temp in temperature:
        k1.append(calculate_six_parameter_fit(reaction,dictonary1,temp))
        k2.append(calculate_six_parameter_fit(reaction,dictonary2,temp))
        gas.TPX = temp,101325,{'Ar':1}
        try:
            k_hong_cti.append(gas.forward_rate_constants[gas.reaction_equations().index(reaction)])
        except:
            pass
        
    plt.figure()
    plt.title(reaction)
    plt.xlabel('Temperature K')
    plt.ylabel('mol/cm^3-s')
    plt.semilogy(temperature,k1,label='MSI Original')
    plt.semilogy(temperature,k2,label='MSI new')
    if reaction in df.columns:
        plt.scatter(df['Temperature'],df[reaction])
        plt.semilogy(temperature,np.array(k_hong_cti)*1000,label='Hong cti')
    plt.legend()
    
    k1 = np.array(k1)
    k2 = np.array(k2)
    difference = k1-k2
    
    percent_difference = ((k1-k2)/k1)*100
    plt.figure()
    plt.plot(temperature,percent_difference)
    plt.title(reaction)
    plt.ylabel('% Difference')
    plt.xlabel('Temperature K')
    #plt.savefig('/home/carly/Dropbox/Columbia/MSI/data/klip_optimization/overlapping_plots_extra_hong_h2o2_data/'+reaction+'percent_difference_'+'.pdf',bbox_inches='tight',dpi=1000)
    
    
```
