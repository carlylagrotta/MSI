```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_raw_data = pd.read_csv('plots_for_ron/raw_hong_data_for_comparing.csv')
df_cantera_generated = pd.read_csv('plots_for_ron/time_history_cantera_generated.csv')
```

```python
plt.plot(df_cantera_generated['time']*1e3,df_cantera_generated['OH']*1e6,label='Cantera Generated Data')
plt.plot(df_raw_data['time_OH']*1e3,df_raw_data['OH_ppm'],label='Graph-read Raw Data')
plt.legend()
plt.title('OH')
plt.xlabel('Time (ms)')
plt.ylabel('OH ppm')
plt.savefig('plots_for_ron/OH_ppm_1.pdf',bbox_inches='tight',dpi=1000)



interpolated_data = np.interp(df_raw_data['time_OH'].dropna().values*1e3,df_cantera_generated['time'].dropna().values*1e3,df_cantera_generated['OH'].dropna().values*1e6)
absolute_difference = interpolated_data-df_raw_data['OH_ppm'].dropna().values
percent_difference = ((interpolated_data-df_raw_data['OH_ppm'].dropna().values)/interpolated_data)*100


plt.figure()
plt.plot(df_raw_data['time_OH'].dropna().values*1e3,absolute_difference)
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')
plt.savefig('plots_for_ron/OH_ppm_absolute_difference_1.pdf',bbox_inches='tight',dpi=1000)


plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df_raw_data['time_OH'].dropna().values*1e3,percent_difference)
plt.savefig('plots_for_ron/OH_ppm_percent_difference_1.pdf',bbox_inches='tight',dpi=1000)
```

```python
#calculating y vecors 
Y = np.log(df_raw_data['OH_ppm'].dropna().values)-np.log(interpolated_data)
Y = Y.reshape((Y.shape[0],1))

length_of_data = Y.shape[0]
time_dependent_uncertainty = np.array([.05]*length_of_data)
z = np.divide(time_dependent_uncertainty,(1/length_of_data**.5))
z = z.reshape((z.shape[0],1))
y = np.divide(Y,z)
y = y.reshape((y.shape[0],1))
```

```python
plt.plot(df_cantera_generated['time']*1e3,df_cantera_generated['H2O']*1e6,label='Cantera Generated Data')
plt.plot(df_raw_data['time_H2O']*1e3,df_raw_data['H2O_ppm'],label='Graph-read Raw Data')
plt.title(r'H$_2$O')
plt.xlabel('Time (ms)')
plt.ylabel(r'H$_2$O ppm')
plt.legend()
plt.savefig('plots_for_ron/H2O_ppm_1.pdf',bbox_inches='tight',dpi=1000)



interpolated_data = np.interp(df_raw_data['time_H2O'].dropna().values*1e3,df_cantera_generated['time'].dropna().values*1e3,df_cantera_generated['H2O'].dropna().values*1e6)
absolute_difference = interpolated_data-df_raw_data['H2O_ppm'].dropna().values
percent_difference = ((interpolated_data-df_raw_data['H2O_ppm'].dropna().values)/interpolated_data)*100


plt.figure()
plt.plot(df_raw_data['time_H2O'].dropna().values*1e3,absolute_difference)
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')
plt.savefig('plots_for_ron/H2O_ppm_absolute_difference_1.pdf',bbox_inches='tight',dpi=1000)


plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df_raw_data['time_H2O'].dropna().values*1e3,percent_difference)
plt.savefig('plots_for_ron/H2O_ppm_percent_difference_1.pdf',bbox_inches='tight',dpi=1000)
```

```python
#calculating y vecors 
Y2 = np.log(df_raw_data['H2O_ppm'].dropna().values)-np.log(interpolated_data)
Y2 = Y2.reshape((Y2.shape[0],1))

length_of_data = Y2.shape[0]
time_dependent_uncertainty = np.array([.05]*length_of_data)
z2 = np.divide(time_dependent_uncertainty,(1/length_of_data**.5))
z2 = z2.reshape((z2.shape[0],1))
y2 = np.divide(Y2,z2)
y2 = y2.reshape((y2.shape[0],1))
```

```python
plt.plot(df_cantera_generated['time']*1e3,df_cantera_generated['Absorbance'],label='Cantera Generated Data')
plt.plot(df_raw_data['time_Abs']*1e3,df_raw_data['Abs'],label='Graph-read Raw Data')
plt.title('Absorbance 227 nm')
plt.xlabel('Time (ms)')
plt.ylabel('Absorbance')
plt.legend()
plt.savefig('plots_for_ron/absorbance_1.pdf',bbox_inches='tight',dpi=1000)



interpolated_data = np.interp(df_raw_data['time_Abs'].dropna().values*1e3,df_cantera_generated['time'].dropna().values*1e3,df_cantera_generated['Absorbance'].dropna().values)
absolute_difference = interpolated_data-df_raw_data['Abs'].dropna().values
percent_difference = ((interpolated_data-df_raw_data['Abs'].dropna().values)/interpolated_data)*100


plt.figure()
plt.plot(df_raw_data['time_Abs'].dropna().values*1e3,absolute_difference)
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')
plt.savefig('plots_for_ron/absorbance_absolute_difference_1.pdf',bbox_inches='tight',dpi=1000)


plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df_raw_data['time_Abs'].dropna().values*1e3,percent_difference)
plt.savefig('plots_for_ron/absorbance_percent_difference_1.pdf',bbox_inches='tight',dpi=1000)
```
