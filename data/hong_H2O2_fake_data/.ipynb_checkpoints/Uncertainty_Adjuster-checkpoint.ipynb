```python
import pandas as pd
import numpy as np
def uncertainty_adjuster(W):
    for csv_file in range(16):
        df = pd.read_csv('hong_h2o_fake_data_'+str(csv_file)+'.csv')
        shape = df.shape[0]
        new_W = np.array([W]*shape)
        new_W = new_W.reshape((new_W.shape[0],1))
        new_W = pd.DataFrame(new_W,columns=['W'])
        df['W'] = new_W['W']
        df.to_csv('large_uncertainty_csv/hong_h2o_fake_data_large_uncertainty_'+str(csv_file)+'.csv', index = False)

        
        df2 = pd.read_csv('hong_oh_fake_data_'+str(csv_file)+'.csv')
        shape2 = df2.shape[0]
        new_W2 = np.array([W]*shape2)
        new_W2 = new_W2.reshape((new_W2.shape[0],1))
        new_W2 = pd.DataFrame(new_W2,columns=['W'])
        df2['W'] = new_W2['W']
        df2.to_csv('large_uncertainty_csv/hong_oh_fake_data_large_uncertainty_'+str(csv_file)+'.csv', index = False)
        
        
        df3 = pd.read_csv('hong_abs_fake_data_'+str(csv_file)+'.csv')
        shape3 = df3.shape[0]
        new_W3 = np.array([W]*shape3)
        new_W3 = new_W3.reshape((new_W3.shape[0],1))
        new_W3 = pd.DataFrame(new_W3,columns=['W'])
        df3['W'] = new_W3['W']
        df3.to_csv('large_uncertainty_csv/hong_abs_fake_data_large_uncertainty_'+str(csv_file)+'.csv', index = False)     
uncertainty_adjuster(.000000001)
```
