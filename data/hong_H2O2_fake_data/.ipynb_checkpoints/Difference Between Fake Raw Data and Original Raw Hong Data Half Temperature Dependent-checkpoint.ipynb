```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df_oh_raw_data = pd.read_csv('/home/carly/Dropbox/Columbia/MSI/data/test_data/hong_oh_4.csv')
df_h2o_raw_data = pd.read_csv('/home/carly/Dropbox/Columbia/MSI/data/test_data/hong_h2o_4.csv')
df_abs_raw_data = pd.read_csv('/home/carly/Dropbox/Columbia/MSI/data/test_data/hong_abs_4.csv')

df = pd.read_csv('plots_for_ron/raw_hong_data_for_comparing.csv')
df_oh_calculated = pd.read_csv('hong_oh_fake_data_3.csv')
df_h2o_calculated = pd.read_csv('hong_h2o_fake_data_3.csv')
df_abs_calculated = pd.read_csv('hong_abs_fake_data_3.csv')
```

```python
plt.plot(df['Time_OH']*1e3,df['OH'],label='Raw Data')
plt.plot(df_oh_calculated['Time']*1e3,df_oh_calculated['OH_ppm'],label='Calculated With Cantera')
interpolated_data = np.interp(df['Time_OH'],df_oh_calculated['Time'],df_oh_calculated['OH_ppm'])
difference = interpolated_data-df['OH']
plt.title('OH')
plt.xlabel('Time (ms)')
plt.ylabel('OH ppm')
plt.legend()
plt.savefig('plots_for_ron/OH_ppm.pdf',bbox_inches='tight',dpi=1000)
    



percent_difference = difference.values/interpolated_data
plt.figure()
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')


plt.plot(df['Time_OH']*1e3,difference)
plt.savefig('plots_for_ron/OH_absolute_difference.pdf',bbox_inches='tight',dpi=1000)

plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df['Time_OH']*1e3,percent_difference*100)
plt.savefig('plots_for_ron/OH_percent_difference.pdf',bbox_inches='tight',dpi=1000)
```

```python
plt.plot(df['Time_H2O']*1e3,df['H2O'],label='Raw Data')
plt.plot(df_h2o_calculated['Time']*1e3,df_h2o_calculated['H2O_ppm'],label='Calculated With Cantera')
plt.title(r'H$_2$O')
plt.xlabel('Time (ms)')
plt.ylabel(r'H$_2$O ppm')
plt.legend()
plt.savefig('plots_for_ron/H2O_ppm.pdf',bbox_inches='tight',dpi=1000)


interpolated_data = np.interp(df['Time_H2O'],df_h2o_calculated['Time'],df_h2o_calculated['H2O_ppm'])
difference = interpolated_data-df['H2O']
percent_difference = difference.values/interpolated_data
plt.figure()
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')
plt.plot(df['Time_H2O'],difference)
plt.savefig('plots_for_ron/H2O_absolute_difference.pdf',bbox_inches='tight',dpi=1000)



plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df['Time_H2O'],percent_difference*100)
plt.savefig('plots_for_ron/H2O_percent_difference.pdf',bbox_inches='tight',dpi=1000)
```

```python
plt.plot(df['Time_abs']*1e3,df['abs'],label='Raw Data')
plt.plot(df_abs_calculated['time']*1e3,df_abs_calculated['Absorbance_227'],label='Calculated With Cantera')
plt.title('Absorbance 227 nm')
plt.xlabel('Time (ms)')
plt.ylabel('Absorbance')
plt.legend()
plt.savefig('plots_for_ron/absorbance.pdf',bbox_inches='tight',dpi=1000)



interpolated_data = np.interp(df['Time_abs'],df_abs_calculated['time'],df_abs_calculated['Absorbance_227'])



difference = interpolated_data-df['abs']
percent_difference = difference.values/interpolated_data
plt.figure()
plt.plot(df['Time_abs']*1e3,difference)
plt.title('Absolute Difference')
plt.xlabel('Time (ms)')
plt.ylabel('Absolute Difference')
plt.savefig('plots_for_ron/absorbance_absolute_difference.pdf',bbox_inches='tight',dpi=1000)




plt.figure()
plt.title('% Difference')
plt.xlabel('Time (ms)')
plt.ylabel('% Difference')
plt.plot(df['Time_abs']*1e3,percent_difference*100)
plt.savefig('plots_for_ron/absorbance_percent_difference.pdf',bbox_inches='tight',dpi=1000)
```
