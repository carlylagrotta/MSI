```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('klip_graphread_data.csv')
plt.semilogy(df['Temp_klip_ttl'],df['k_klip_ttl'], label='Klip. Ttl Rate Constant',linewidth=7.0)
plt.scatter(df['temp_experimental'],df['experimental_data'],color='k')
plt.semilogy(df['temp_klip'],df['k_klip'],label='Klip. Rate Constant')
plt.semilogy(df['T_original_low_b'],df['k_original_low_b'],'b--',label='')
plt.semilogy(df['T_original_high_b'],df['k_original_high_b'],'b--',label='')
plt.semilogy(df['T_original_b'],df['k_original_b'],'b',label='')

plt.semilogy(df['T_original_low_r'],df['k_original_low_r'],'r--',label='')
plt.semilogy(df['T_original_high_r'],df['k_original_high_r'],'r--',label='')
plt.semilogy(df['T_original_r'],df['k_original_r'],'r',label='')
plt.xlabel('Temperature K')
plt.ylabel('mol/cm^3-s')
plt.title('2HO2 <=> H2O2 + O2')
plt.legend()
plt.savefig('2 HO2 <=> H2O2 + O2'+'.pdf', bbox_inches='tight',dpi=1000)
```
