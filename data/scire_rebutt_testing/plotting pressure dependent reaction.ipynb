```python
import cantera as ct
import matplotlib.pyplot as plt
import numpy as np
gas = ct.Solution('FFCM1_custom.cti')
gas.reaction_equations() 
temp_list = np.arange(200,2400)
p_list = np.arange(1,11)
print(gas.reaction_equations()[0])
plt.figure()
for pressure in p_list:
    k =[]
    for temp in temp_list:
        gas.TPX = temp,pressure*101325,{'H2O2':0.002046 ,'H2O':0.001113,
                                     'O2':0.000556,'Ar':0.995212}
        k.append(gas.forward_rate_constants[0]*1000)
    plt.semilogy(temp_list,k,label='Pressure:'+str(pressure) + ' atm')
    plt.legend(ncol=2)
    
        
```
